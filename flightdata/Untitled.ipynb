```python
str.replace('Via','').split()
```

```python
def stopsfind(stops): # "1-stop\n\t\t\" , "1-stop\n\t\tVia Vishakhapatnam\n\t\t\"    
    stops = stops.replace('Via','').split()
    if len(stops) == 1:
        return pd.Series([stops[0],"nan"])
    else:
        return pd.Series([stops[0],stops[1]])
```
